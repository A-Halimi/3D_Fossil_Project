## FINAL ENSEMBLE WEIGHTED MODEL  ConvNeXt-L  +  Eff-V2-S patch

```python
# ============================================================================
#  FINAL Saving Script
# ============================================================================
import os, pathlib, json, numpy as np, tensorflow as tf, matplotlib.pyplot as plt, seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras import mixed_precision, saving
from tensorflow.keras.saving import register_keras_serializable

# --- (Sections 0, 1, 2, 3 are unchanged) ---
# ░░ 0 · Environment ░░
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"
mixed_precision.set_global_policy("mixed_float16")
# Get the root directory of the project (Fossil_Project)
PROJECT_ROOT = pathlib.Path(__file__).resolve().parent.parent
AUTOTUNE, SEED = tf.data.AUTOTUNE, 42
CANONICAL_SIZE = (384, 384)
DATA_ROOT    = PROJECT_ROOT / "1_Dataset_Creation" / "3d_fossil_dataset_segmented_final"
RESULTS_ROOT = PROJECT_ROOT / "3_Results" # Path to the results directory
SPLIT        = "test"
MAIN, PATCH  = "convnextl", "effv2s"
WEAK_CLASSES = {"Baculogypsina", "Orbitoides"}

# ░░ 1 · Robust bbox crop ░░
@tf.autograph.experimental.do_not_convert
def crop_to_bbox(img):
    gray   = tf.image.rgb_to_grayscale(img)
    coords = tf.cast(tf.where(tf.squeeze(tf.where(gray > 0.05, 1, 0))), tf.int32)
    def _crop():
        ymin, ymax = tf.reduce_min(coords[:,0])-5, tf.reduce_max(coords[:,0])+5
        xmin, xmax = tf.reduce_min(coords[:,1])-5, tf.reduce_max(coords[:,1])+5
        h, w = tf.shape(img, tf.int32)[0], tf.shape(img, tf.int32)[1]
        ymin, ymax = tf.clip_by_value(ymin,0,h), tf.clip_by_value(ymax,0,h)
        xmin, xmax = tf.clip_by_value(xmin,0,w), tf.clip_by_value(xmax,0,w)
        empty = tf.logical_or(tf.less_equal(ymax,ymin), tf.less_equal(xmax,xmin))
        return tf.cond(empty, lambda: img, lambda: img[ymin:ymax, xmin:xmax])
    crop = tf.cond(tf.shape(coords)[0] > 0, _crop, lambda: img)
    return tf.cast(tf.image.resize_with_pad(crop, *CANONICAL_SIZE), img.dtype)

# ░░ 2 · Pre-processing helpers ░░
from tensorflow.keras.applications.convnext      import preprocess_input as cvx_prep
from tensorflow.keras.applications.efficientnet_v2 import preprocess_input as eff_prep
PREP  = {MAIN: cvx_prep, PATCH: eff_prep}
IN_SZ = {MAIN:(384,384), PATCH:(224,224)}

# ░░ 3 · Load checkpoints ░░
def load(tag):
    p = RESULTS_ROOT/tag/"ckpts"/"best.keras"
    if not p.exists(): p = RESULTS_ROOT/tag/f"fossil_{tag}.keras"
    return tf.keras.models.load_model(p, compile=False)
main_m, patch_m = load(MAIN), load(PATCH)

# ░░ 4 · Base dataset (no prep yet) ░░
def base_ds(bs=64):
    raw = tf.keras.utils.image_dataset_from_directory(
        DATA_ROOT/SPLIT, image_size=CANONICAL_SIZE, batch_size=bs,
        label_mode="categorical", shuffle=False, seed=SEED)
    names = raw.class_names
    # >>> The one and only place where crop_to_bbox is applied <<<
    ds = raw.map(lambda x,y:(tf.map_fn(crop_to_bbox, x), y), num_parallel_calls=AUTOTUNE)
    return ds.prefetch(AUTOTUNE), names
base, CLASS_NAMES = base_ds()
weak_idx = np.array([CLASS_NAMES.index(c) for c in WEAK_CLASSES])

# --- (Sections 5 and 6 for evaluation are unchanged) ---
# ░░ 5 · Helper to run a branch ░░
def branch_probs(model, tag):
    prep, sz = PREP[tag], IN_SZ[tag]
    ds = base.map(lambda x,y:(tf.cast(prep(tf.image.resize(x,sz)), tf.float32), y),
                  num_parallel_calls=AUTOTUNE)
    return model.predict(ds, verbose=1)

pm = branch_probs(main_m , MAIN )
pp = branch_probs(patch_m, PATCH)

# ░░ 6 · Confidence-wins rule ░░
top1     = pm.argmax(1)
mask_cls = np.isin(top1, weak_idx)
mask_cmp = pp.max(1) > pm.max(1)
mask     = mask_cls & mask_cmp

probs = pm.copy()
probs[mask] = pp[mask]

yt = np.concatenate([y for _,y in base]).argmax(1)
yp = probs.argmax(1)

print("\n--- Patched Ensemble Report ---")
print(classification_report(yt, yp, target_names=CLASS_NAMES, digits=4))

cm = confusion_matrix(yt, yp)
plt.figure(figsize=(14,12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
plt.title("Confusion Matrix – ConvNeXt-L  +  Eff-V2-S patch")
plt.tight_layout(); plt.show()

# ============================================================================
# ░░ 7 · CORRECTED DEPLOYABLE MODEL ░░
# ============================================================================

@register_keras_serializable(package="Fossil")
class PatchEnsemble(tf.keras.Model):
    def __init__(self, main, patch, weak_ids, **kw):
        super().__init__(**kw)
        self.main = main
        self.patch = patch
        self.weak = tf.constant(weak_ids, dtype=tf.int64)
        self.m_prep = tf.keras.layers.Lambda(cvx_prep, name="main_prep")
        self.p_prep = tf.keras.layers.Lambda(eff_prep, name="patch_prep")

    # It assumes the input `x` has already been cropped and padded.
    def call(self, x, training=False):
        # The redundant `crop_to_bbox` call has been REMOVED.
        pm = self.main(self.m_prep(x), training=False)
        pp = self.patch(self.p_prep(tf.image.resize(x, IN_SZ[PATCH])), training=False)

        top1 = tf.argmax(pm, -1, output_type=tf.int64)
        better = tf.reduce_max(pp,-1) > tf.reduce_max(pm,-1)
        weak   = tf.reduce_any(top1[:,None]==self.weak[None,:], axis=1)
        mask   = tf.logical_and(weak, better)
        mask_f = tf.cast(mask[:,None], pm.dtype)
        out = pm*(1-mask_f) + pp*mask_f
        return out / tf.reduce_sum(out, -1, keepdims=True)

    def get_config(self):
        base_config = super().get_config()
        config = { "main": saving.serialize_keras_object(self.main), "patch": saving.serialize_keras_object(self.patch), "weak_ids": self.weak.numpy().tolist() }
        return {**base_config, **config}

    @classmethod
    def from_config(cls, config):
        main_model = saving.deserialize_keras_object(config.pop("main"))
        patch_model = saving.deserialize_keras_object(config.pop("patch"))
        return cls(main=main_model, patch=patch_model, **config)

# --- Create and Save the Corrected Deployable Model ---
patch_net = PatchEnsemble(main_m, patch_m, weak_idx)
_ = patch_net(tf.zeros([1, *CANONICAL_SIZE, 3]))

DEPLOY_DIR = RESULTS_ROOT / "fossil_classifier_final" # Saving to a new directory
DEPLOY_DIR.mkdir(exist_ok=True)
DEPLOY_MODEL_FILE = DEPLOY_DIR / "model.keras"
DEPLOY_CLASSES_FILE = DEPLOY_DIR / "class_names.json"

patch_net.save(DEPLOY_MODEL_FILE)
with open(DEPLOY_CLASSES_FILE, 'w') as f:
    json.dump(CLASS_NAMES, f)

print(f"\n✅ Saved CORRECTED model and assets to: {DEPLOY_DIR}")
```

## LOAD AND PREDICT 

```python
# ============================================================================
#  FINAL Prediction Script  - User-Controlled CLAHE Normalization
# ============================================================================
import os, json, pathlib, numpy as np, tensorflow as tf, cv2
from tensorflow.keras import mixed_precision, saving
from tensorflow.keras.saving import register_keras_serializable
from skimage import filters, morphology, measure
import matplotlib.pyplot as plt
import sys

# --- Setup and Custom Object Definitions (Unchanged) ---
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"; os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"
mixed_precision.set_global_policy("mixed_float16")
CANONICAL_SIZE = (384, 384); MAIN, PATCH = "convnextl", "effv2s"; IN_SZ = {MAIN:(384,384), PATCH:(224,224)}
from tensorflow.keras.applications.convnext import preprocess_input as cvx_prep
from tensorflow.keras.applications.efficientnet_v2 import preprocess_input as eff_prep

@tf.autograph.experimental.do_not_convert
def crop_to_bbox(img):
    gray=tf.image.rgb_to_grayscale(img);coords=tf.cast(tf.where(tf.squeeze(tf.where(gray>0.05,1,0))),tf.int32)
    def _crop():
        ymin,ymax=tf.reduce_min(coords[:,0])-5,tf.reduce_max(coords[:,0])+5;xmin,xmax=tf.reduce_min(coords[:,1])-5,tf.reduce_max(coords[:,1])+5
        h,w=tf.shape(img,tf.int32)[0],tf.shape(img,tf.int32)[1];ymin,ymax=tf.clip_by_value(ymin,0,h),tf.clip_by_value(ymax,0,h);xmin,xmax=tf.clip_by_value(xmin,0,w),tf.clip_by_value(xmax,0,w)
        empty=tf.logical_or(tf.less_equal(ymax,ymin),tf.less_equal(xmax,xmin));return tf.cond(empty,lambda:img,lambda:img[ymin:ymax,xmin:xmax])
    crop=tf.cond(tf.shape(coords)[0]>0,_crop,lambda:img);return tf.cast(tf.image.resize_with_pad(crop,*CANONICAL_SIZE),img.dtype)
@register_keras_serializable(package="Fossil")
class PatchEnsemble(tf.keras.Model):
    def __init__(self,main,patch,weak_ids,**kw):super().__init__(**kw);self.main,self.patch,self.weak=main,patch,tf.constant(weak_ids,dtype=tf.int64);self.m_prep=tf.keras.layers.Lambda(cvx_prep,name="main_prep");self.p_prep=tf.keras.layers.Lambda(eff_prep,name="patch_prep")
    def call(self,x,training=False):pm=self.main(self.m_prep(x),training=False);pp=self.patch(self.p_prep(tf.image.resize(x,IN_SZ[PATCH])),training=False);top1=tf.argmax(pm,-1,output_type=tf.int64);better=tf.reduce_max(pp,-1)>tf.reduce_max(pm,-1);weak=tf.reduce_any(top1[:,None]==self.weak[None,:],axis=1);mask=tf.logical_and(weak,better);mask_f=tf.cast(mask[:,None],pm.dtype);out=pm*(1-mask_f)+pp*mask_f;return out/tf.reduce_sum(out,-1,keepdims=True)
    def get_config(self):base_config=super().get_config();config={"main":saving.serialize_keras_object(self.main),"patch":saving.serialize_keras_object(self.patch),"weak_ids":self.weak.numpy().tolist()};return {**base_config,**config}
    @classmethod
    def from_config(cls,config):main_model=saving.deserialize_keras_object(config.pop("main"));patch_model=saving.deserialize_keras_object(config.pop("patch"));return cls(main=main_model,patch=patch_model,**config)

def segment_image(image_path,assume_bright_fossil=True,invert_output=False):
    try:
        img=cv2.imread(str(image_path),cv2.IMREAD_GRAYSCALE);
        if img is None:return None
        blurred=cv2.GaussianBlur(img,(3,3),0);threshold=filters.threshold_otsu(blurred);binary=blurred>threshold if assume_bright_fossil else blurred<threshold
        cleaned=morphology.remove_small_objects(binary,min_size=200);cleaned=morphology.remove_small_holes(cleaned,area_threshold=100);cleaned=morphology.binary_closing(cleaned,morphology.disk(2))
        labeled=measure.label(cleaned)
        if labeled.max()==0:return None
        main_fossil_label=max(measure.regionprops(labeled),key=lambda x:x.area).label;mask=(labeled==main_fossil_label).astype(np.uint8)*255
        segmented_gray=cv2.bitwise_and(img,img,mask=mask)
        if invert_output:fossil_pixels=segmented_gray[mask==255];segmented_gray[mask==255]=255-fossil_pixels
        return segmented_gray
    except Exception as e:print(f"Error segmenting {image_path}: {e}");return None

def predict_on_new_images():
    # ==================== CONTROL PANEL ====================
    # Set to True for maximum robustness on mixed-quality images.
    # Set to False for maximum confidence on "perfect" grayscale images.
    USE_CLAHE_NORMALIZATION = False
    # =======================================================
    # Get the root directory of the project (Fossil_Project)
    try:
        # Works in normal .py scripts
        PROJECT_ROOT = pathlib.Path(__file__).resolve().parent.parent
    except NameError:
        # Fallback for Jupyter notebooks
        PROJECT_ROOT = pathlib.Path.cwd().parent

    DEPLOY_DIR=PROJECT_ROOT / "3_Results" / "fossil_classifier_final"
    MODEL_PATH=DEPLOY_DIR/"model.keras"
    CLASSES_PATH=DEPLOY_DIR/"class_names.json"
    # Path to the Images_samples folder
    IMAGES_DIR = pathlib.Path("Image_samples")

    # Build full paths
    image_paths = [
        IMAGES_DIR / "avizo_Chrysalidina.png",
        IMAGES_DIR / "avizo_Orbitoides.png",
        IMAGES_DIR / "avizo_Fallotia.png",
        IMAGES_DIR / "avizo_Lockhartia.png",
        IMAGES_DIR / "avizo_Alveolina.png",
        IMAGES_DIR / "Arumella_specimen2.png"
    ]



    if not MODEL_PATH.exists():print(f"[ERROR] Model file not found at: {MODEL_PATH}");return
    with open(CLASSES_PATH,'r')as f:CLASS_NAMES=json.load(f)
    model=tf.keras.models.load_model(MODEL_PATH)
    clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
    prediction_data=[]
    print("\n[*] Preprocessing images...")
    for path in image_paths:
        original_img_bgr=cv2.imread(str(path))
        if original_img_bgr is None:print(f"Warning: Could not read original image at {path}, skipping.");continue
        img_gray_for_check=cv2.cvtColor(original_img_bgr,cv2.COLOR_BGR2GRAY)
        if np.mean(img_gray_for_check)>190:
            segmented_np=segment_image(path,assume_bright_fossil=False,invert_output=True)
        else:
            segmented_np=segment_image(path,assume_bright_fossil=True,invert_output=False)
        if segmented_np is not None:
            processed_np = segmented_np
            if USE_CLAHE_NORMALIZATION:
                print(f"  -> Applying CLAHE normalization to {os.path.basename(path)}")
                processed_np = clahe.apply(segmented_np)
            final_rgb_np=cv2.cvtColor(processed_np,cv2.COLOR_GRAY2RGB)
            img_tensor=tf.convert_to_tensor(final_rgb_np,dtype=tf.float32);cropped_tensor=crop_to_bbox(img_tensor)
            prediction_data.append({"path":path,"original_image":cv2.cvtColor(original_img_bgr,cv2.COLOR_BGR2RGB),"processed_tensor":cropped_tensor})
        else:
            print(f"Warning: Segmentation failed for {path}. Skipping.")
    if not prediction_data:print("[ERROR] Preprocessing failed for all images.");return
    image_batch=tf.stack([item["processed_tensor"] for item in prediction_data])
    print("\n[*] Running predictions...")
    predictions=model.predict(image_batch)
    print("\n"+"="*60+"\n                      PREDICTION RESULTS\n"+"="*60)
    for i,data in enumerate(prediction_data):
        path,original_image=data["path"],data["original_image"];final_processed_image=image_batch[i].numpy().astype("uint8");probs=predictions[i];top_5_indices=np.argsort(probs)[-5:][::-1]
        print(f"\n--- Results for: {os.path.basename(path)} ---\n\n🏆 Winner: {CLASS_NAMES[top_5_indices[0]]} (Confidence: {probs[top_5_indices[0]]:.4f})\n\nTop 5 Predictions:")
        for j in top_5_indices:print(f"  - {CLASS_NAMES[j]:<20} (Confidence: {probs[j]:.4f})")
        # Add extra blank lines before the figure
        print("\n" * 1)
        fig,axes=plt.subplots(1,2,figsize=(10,5));axes[0].imshow(original_image);axes[0].set_title("Original Uploaded Image");axes[0].axis('off');axes[1].imshow(final_processed_image,cmap='gray');axes[1].set_title("Image Fed to Model");axes[1].axis('off');fig.suptitle(f"Prediction for {os.path.basename(path)}",fontsize=16,y=0.97);plt.show()
        print("\n" * 1)
    print("\n"+"="*60)

if __name__ == "__main__":
    predict_on_new_images()
```

```python
# run_prediction.py

import os
import fossil_predictor  # Import the library you just created
from pprint import pprint
import pathlib
import sys

def main():
    """
    Main function to run fossil predictions using the fossil_predictor library.
    """
    # --- Configuration ---
    # Directory where the 'model.keras' and 'class_names.json' files are located
    try:
        # Works in normal .py scripts
        PROJECT_ROOT = pathlib.Path(__file__).resolve().parent.parent
    except NameError:
        # Fallback for Jupyter notebooks
        PROJECT_ROOT = pathlib.Path.cwd().parent
        
    MODEL_DIRECTORY=PROJECT_ROOT / "3_Results" / "fossil_classifier_final"

    
    # List of image files to predict on
    IMAGES_DIR = pathlib.Path("Image_samples")

    # Build full paths
    image_paths = [
        IMAGES_DIR / "avizo_Chrysalidina.png",
        IMAGES_DIR / "avizo_Orbitoides.png",
        IMAGES_DIR / "avizo_Fallotia.png",
        IMAGES_DIR / "avizo_Lockhartia.png",
        IMAGES_DIR / "avizo_Alveolina.png",
        IMAGES_DIR / "Arumella_specimen2.png"
    ]
        
    # --- Check if files exist ---
    if not os.path.isdir(MODEL_DIRECTORY):
        print(f"Error: Model directory not found at '{MODEL_DIRECTORY}'")
        return
    for img_path in IMAGE_SAMPLES:
        if not os.path.exists(img_path):
            print(f"Error: Image sample not found at '{img_path}'")
            return
            
    # --- Run Prediction ---
    # Call the predict function from your library.
    # We set show_plots=False to only get the data back.
    prediction_results = fossil_predictor.predict(
        image_paths=IMAGE_SAMPLES,
        model_dir=MODEL_DIRECTORY,
        use_clahe=False,  # Use True for robustness, False for higher confidence on perfect images
        show_plots=False # Set to True if you want to see the plots here
    )
    
    # --- Display Results ---
    print("\n" + "="*60)
    print("               PREDICTION RESULTS FROM SCRIPT")
    print("="*60)
    
    if prediction_results:
        # pprint is a "pretty-print" function that makes dictionaries easier to read
        pprint(prediction_results)
    else:
        print("No results were returned from the predictor.")

if __name__ == "__main__":
    main()
```
