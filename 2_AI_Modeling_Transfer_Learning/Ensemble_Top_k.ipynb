```python
#!/usr/bin/env python
# ============================================================================
#  Fossil Ensemble · Top-4 branches  ·  bbox-crop  ·  NaN-safe class weights
# ============================================================================

import os, gc, pathlib, numpy as np, pandas as pd, tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt, seaborn as sns
from keras.saving import register_keras_serializable
from tensorflow.keras import mixed_precision

# ░░ 0 · Environment ░░
os.environ.setdefault("TF_CPP_MIN_LOG_LEVEL", "2")
os.environ.setdefault("TF_FORCE_GPU_ALLOW_GROWTH", "true")
mixed_precision.set_global_policy("mixed_float16")
AUTOTUNE, SEED = tf.data.AUTOTUNE, 42
CANONICAL_SIZE = (384, 384)

# Get the root directory of the project (Fossil_Project)
try:
    # Works in normal .py scripts
    PROJECT_ROOT = pathlib.Path(__file__).resolve().parent.parent
except NameError:
    # Fallback for Jupyter notebooks
    PROJECT_ROOT = pathlib.Path.cwd().parent

# ░░ 1 · User params ░░
DATA_ROOT = PROJECT_ROOT / "1_Dataset_Creation" / "3d_fossil_dataset_segmented_final"
RESULTS_ROOT = PROJECT_ROOT / "3_Results"
CSV_OVERALL  = RESULTS_ROOT / "3_Results" / "_comparison/fossil_model_comparison_report_v2.csv"
CSV_PERCLS   = RESULTS_ROOT / "3_Results" / "_comparison/fossil_model_per_class_metrics_v2.csv"
DEPLOY_FILE  = RESULTS_ROOT / "3_Results" / "deployable_fossil_ensemble_segmented_top4.keras"
SPLIT = "test"
TOP_K = 4                                         # keep best 4 models

# ░░ 1 · Bounding-box crop ░░
@tf.autograph.experimental.do_not_convert
def crop_to_bbox(img):
    gray  = tf.image.rgb_to_grayscale(img)
    mask  = tf.where(gray > 0.05, 1, 0)
    coords = tf.cast(tf.where(tf.squeeze(mask)), tf.int32)

    def _do_crop():
        ymin = tf.reduce_min(coords[:,0])-5; ymax = tf.reduce_max(coords[:,0])+5
        xmin = tf.reduce_min(coords[:,1])-5; xmax = tf.reduce_max(coords[:,1])+5
        h = tf.shape(img, tf.int32)[0]; w = tf.shape(img, tf.int32)[1]
        ymin = tf.clip_by_value(ymin,0,h); ymax = tf.clip_by_value(ymax,0,h)
        xmin = tf.clip_by_value(xmin,0,w); xmax = tf.clip_by_value(xmax,0,w)
        return img[ymin:ymax, xmin:xmax]

    crop = tf.cond(tf.shape(coords)[0] > 0, _do_crop, lambda: img)
    out  = tf.image.resize_with_pad(crop, *CANONICAL_SIZE)
    return tf.cast(out, img.dtype)

# ░░ 2 · Registry (same as before) ░░
def _torch_prep(x):
    from tensorflow.keras.applications.convnext import preprocess_input as p
    return p(x/255.0)

def _caffe_prep(name):
    def f(x):
        if name == "mobilenet":
            from tensorflow.keras.applications.mobilenet import preprocess_input as p
        elif name in {"effv2s","effv2l"}:
            from tensorflow.keras.applications.efficientnet_v2 import preprocess_input as p
        elif name == "nasnet":
            from tensorflow.keras.applications.nasnet import preprocess_input as p
        else:
            from tensorflow.keras.applications.resnet_v2 import preprocess_input as p
        return p(x)
    return f

REG = {
    "mobilenet":   (_caffe_prep("mobilenet"),   (224,224)),
    "effv2s":      (_caffe_prep("effv2s"),      (224,224)),
    "effv2l":      (_caffe_prep("effv2l"),      (384,384)),
    "nasnet":      (_caffe_prep("nasnet"),      (331,331)),
    "resnet101v2": (_caffe_prep("resnet101v2"), (224,224)),
    "convnext":    (_torch_prep,                (224,224)),
    "convnextl":   (_torch_prep,                (384,384)),
}

# ░░ 3 · Choose top-K models ░░
overall = pd.read_csv(CSV_OVERALL)
overall["model"] = overall["model"].str.strip()
TOP_MODELS = (overall.sort_values("macro_f1", ascending=False)
                     .head(TOP_K)["model"].tolist())
print("Top-K models:", TOP_MODELS)

# ░░ 4 · Load per-class metrics ░░
percls = pd.read_csv(CSV_PERCLS, header=[0,1], index_col=0)
percls.columns = percls.columns.set_levels(
    percls.columns.levels[0].str.strip(), level=0)

# ░░ 5 · Weight builders ░░
def global_w(models):
    v = overall.set_index("model").loc[models]["macro_f1"].astype("float32").values
    v = np.maximum(v,1e-8)
    return v / v.sum()

def class_w(models, classes, mix=0.5):     #0.35
    C,M = len(classes), len(models)
    W = np.zeros((C,M), dtype="float32")
    for ci, cls in enumerate(classes):
        for mi, m in enumerate(models):
            val = percls.loc[cls,(m,"f1")] if (cls,(m,"f1")) in percls else np.nan
            if np.isnan(val):
                val = percls.loc[cls,(m,"recall")] if (cls,(m,"recall")) in percls else 0.0
            W[ci,mi] = max(0.0, float(val))
        row_sum = W[ci].sum()
        if row_sum == 0:
            W[ci] = 1.0                                        # flat before mix
        else:
            W[ci] /= row_sum
    g = global_w(models)
    W = (1-mix)*W + mix*g[np.newaxis,:]
    W /= W.sum(1, keepdims=True)
    return W

# ░░ 6 · Data helpers ░░
def batch_for(size, base_px=224*224, base_bs=64):
    return max(1, int(round(base_bs*(base_px/(size[0]*size[1])))))

def build_ds(root,size,prep,bs):
    raw = tf.keras.utils.image_dataset_from_directory(
        root/SPLIT, image_size=CANONICAL_SIZE, batch_size=bs,
        label_mode="categorical", shuffle=False, seed=SEED)
    names = raw.class_names
    ds = raw.map(lambda x,y:(tf.map_fn(crop_to_bbox,x), y), num_parallel_calls=AUTOTUNE)
    ds = ds.map(lambda x,y:(prep(tf.image.resize(x,size)), y), num_parallel_calls=AUTOTUNE)
    return ds.prefetch(AUTOTUNE), names

# ░░ 7 · Inference ░░
probs_list, y_true, class_names, branches = [], None, None, []
for m in TOP_MODELS:
    print(f"\n==> Inferring with member: {m}")
    prep,size = REG[m]
    print(f"    using batch size: {batch_for(size)}")
    ds, class_names = build_ds(DATA_ROOT, size, prep, batch_for(size))
    if y_true is None:
        y_true = np.concatenate([y for _,y in ds])
    path = RESULTS_ROOT/m/"ckpts"/"best.keras"
    if not path.exists():
        path = RESULTS_ROOT/m/f"fossil_{m}.keras"
    model = tf.keras.models.load_model(path, compile=False)
    probs_list.append(model.predict(ds, verbose=1).astype("float32"))
    branches.append((model,size,prep))
    gc.collect()

probs = np.stack(probs_list, -1)            # (N,C,M)
W = class_w(TOP_MODELS, class_names, mix=0.35)
ens = (probs * W[np.newaxis,:,:]).sum(-1)
ens /= ens.sum(1, keepdims=True)

# ░░ 8 · Report ░░
yt, yp = y_true.argmax(1), ens.argmax(1)
print("\n--- Class-Aware Weighted Ensemble Report (In-Memory) ---")
print(classification_report(yt, yp, target_names=class_names, digits=4))

report_str = classification_report(yt, yp, target_names=class_names, digits=4)

out_path = RESULTS_ROOT / "classification_report_ensemble_top-4.txt"
out_path.parent.mkdir(parents=True, exist_ok=True)  # just in case

with open(out_path, "w", encoding="utf-8") as f:
    f.write("--- Class-Aware Weighted Ensemble Report ---\n")
    f.write(report_str + "\n")

cm = confusion_matrix(yt, yp)
plt.figure(figsize=(14,12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
plt.title("Confusion Matrix – Top-4 Ensemble")
plt.tight_layout(); plt.show()

# ░░ 9 · Deployable model ░░
@register_keras_serializable(package="Fossil")
class ClassAwareEnsemble(tf.keras.Model):
    def __init__(self,names,branches,sizes,W,**kw):
        super().__init__(**kw)
        self.br   = [b for b,_,_ in branches]
        self.sz   = sizes
        self.prep = [tf.keras.layers.Lambda(p) for _,_,p in branches]
        self.W    = tf.Variable(W, trainable=False, dtype="float32")

    def call(self,x,training=False):
        x = tf.map_fn(crop_to_bbox, x)
        x = tf.image.grayscale_to_rgb(tf.image.rgb_to_grayscale(tf.cast(x,self.compute_dtype)))
        outs = [m(prep(tf.image.resize(x, sz)), training=False)
                for m, sz, prep in zip(self.br, self.sz, self.prep)]
        y = tf.stack(outs, -1) * tf.cast(self.W, outs[0].dtype)[tf.newaxis,:,:]
        y = tf.reduce_sum(y, -1)
        return y / tf.reduce_sum(y, -1, keepdims=True)

sizes = [s for _,s,_ in branches]
ensemble = ClassAwareEnsemble(TOP_MODELS, branches, sizes, tf.constant(W))
_ = ensemble(tf.zeros([1,*CANONICAL_SIZE,3]))
ensemble.save(DEPLOY_FILE)
print("\n✅ Saved top-4 ensemble to", DEPLOY_FILE)
```

```python
#!/usr/bin/env python
# ============================================================================
#  Fossil Ensemble · Top-4 branches  ·  bbox-crop  ·  NaN-safe class weights
# ============================================================================

import os, gc, pathlib, numpy as np, pandas as pd, tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt, seaborn as sns
from keras.saving import register_keras_serializable
from tensorflow.keras import mixed_precision

# ░░ 0 · Environment ░░
os.environ.setdefault("TF_CPP_MIN_LOG_LEVEL", "2")
os.environ.setdefault("TF_FORCE_GPU_ALLOW_GROWTH", "true")
mixed_precision.set_global_policy("mixed_float16")
AUTOTUNE, SEED = tf.data.AUTOTUNE, 42
CANONICAL_SIZE = (384, 384)

# Get the root directory of the project (Fossil_Project)
try:
    # Works in normal .py scripts
    PROJECT_ROOT = pathlib.Path(__file__).resolve().parent.parent
except NameError:
    # Fallback for Jupyter notebooks
    PROJECT_ROOT = pathlib.Path.cwd().parent

# ░░ 1 · User params ░░
DATA_ROOT = PROJECT_ROOT / "1_Dataset_Creation" / "3d_fossil_dataset_segmented_final"
RESULTS_ROOT = PROJECT_ROOT / "3_Results"
CSV_OVERALL  = RESULTS_ROOT / "3_Results" / "_comparison/fossil_model_comparison_report_v2.csv"
CSV_PERCLS   = RESULTS_ROOT / "3_Results" / "_comparison/fossil_model_per_class_metrics_v2.csv"
DEPLOY_FILE  = RESULTS_ROOT / "3_Results" / "deployable_fossil_ensemble_segmented_top4.keras"
SPLIT = "test"
TOP_K = 2                                         # keep best 4 models

# ░░ 1 · Bounding-box crop ░░
@tf.autograph.experimental.do_not_convert
def crop_to_bbox(img):
    gray  = tf.image.rgb_to_grayscale(img)
    mask  = tf.where(gray > 0.05, 1, 0)
    coords = tf.cast(tf.where(tf.squeeze(mask)), tf.int32)

    def _do_crop():
        ymin = tf.reduce_min(coords[:,0])-5; ymax = tf.reduce_max(coords[:,0])+5
        xmin = tf.reduce_min(coords[:,1])-5; xmax = tf.reduce_max(coords[:,1])+5
        h = tf.shape(img, tf.int32)[0]; w = tf.shape(img, tf.int32)[1]
        ymin = tf.clip_by_value(ymin,0,h); ymax = tf.clip_by_value(ymax,0,h)
        xmin = tf.clip_by_value(xmin,0,w); xmax = tf.clip_by_value(xmax,0,w)
        return img[ymin:ymax, xmin:xmax]

    crop = tf.cond(tf.shape(coords)[0] > 0, _do_crop, lambda: img)
    out  = tf.image.resize_with_pad(crop, *CANONICAL_SIZE)
    return tf.cast(out, img.dtype)

# ░░ 2 · Registry (same as before) ░░
def _torch_prep(x):
    from tensorflow.keras.applications.convnext import preprocess_input as p
    return p(x/255.0)

def _caffe_prep(name):
    def f(x):
        if name == "mobilenet":
            from tensorflow.keras.applications.mobilenet import preprocess_input as p
        elif name in {"effv2s","effv2l"}:
            from tensorflow.keras.applications.efficientnet_v2 import preprocess_input as p
        elif name == "nasnet":
            from tensorflow.keras.applications.nasnet import preprocess_input as p
        else:
            from tensorflow.keras.applications.resnet_v2 import preprocess_input as p
        return p(x)
    return f

REG = {
    "mobilenet":   (_caffe_prep("mobilenet"),   (224,224)),
    "effv2s":      (_caffe_prep("effv2s"),      (224,224)),
    "effv2l":      (_caffe_prep("effv2l"),      (384,384)),
    "nasnet":      (_caffe_prep("nasnet"),      (331,331)),
    "resnet101v2": (_caffe_prep("resnet101v2"), (224,224)),
    "convnext":    (_torch_prep,                (224,224)),
    "convnextl":   (_torch_prep,                (384,384)),
}

# ░░ 3 · Choose top-K models ░░
overall = pd.read_csv(CSV_OVERALL)
overall["model"] = overall["model"].str.strip()
TOP_MODELS = (overall.sort_values("macro_f1", ascending=False)
                     .head(TOP_K)["model"].tolist())
print("Top-K models:", TOP_MODELS)

# ░░ 4 · Load per-class metrics ░░
percls = pd.read_csv(CSV_PERCLS, header=[0,1], index_col=0)
percls.columns = percls.columns.set_levels(
    percls.columns.levels[0].str.strip(), level=0)

# ░░ 5 · Weight builders ░░
def global_w(models):
    v = overall.set_index("model").loc[models]["macro_f1"].astype("float32").values
    v = np.maximum(v,1e-8)
    return v / v.sum()

def class_w(models, classes, mix=0.5):     #0.35
    C,M = len(classes), len(models)
    W = np.zeros((C,M), dtype="float32")
    for ci, cls in enumerate(classes):
        for mi, m in enumerate(models):
            val = percls.loc[cls,(m,"f1")] if (cls,(m,"f1")) in percls else np.nan
            if np.isnan(val):
                val = percls.loc[cls,(m,"recall")] if (cls,(m,"recall")) in percls else 0.0
            W[ci,mi] = max(0.0, float(val))
        row_sum = W[ci].sum()
        if row_sum == 0:
            W[ci] = 1.0                                        # flat before mix
        else:
            W[ci] /= row_sum
    g = global_w(models)
    W = (1-mix)*W + mix*g[np.newaxis,:]
    W /= W.sum(1, keepdims=True)
    return W

# ░░ 6 · Data helpers ░░
def batch_for(size, base_px=224*224, base_bs=64):
    return max(1, int(round(base_bs*(base_px/(size[0]*size[1])))))

def build_ds(root,size,prep,bs):
    raw = tf.keras.utils.image_dataset_from_directory(
        root/SPLIT, image_size=CANONICAL_SIZE, batch_size=bs,
        label_mode="categorical", shuffle=False, seed=SEED)
    names = raw.class_names
    ds = raw.map(lambda x,y:(tf.map_fn(crop_to_bbox,x), y), num_parallel_calls=AUTOTUNE)
    ds = ds.map(lambda x,y:(prep(tf.image.resize(x,size)), y), num_parallel_calls=AUTOTUNE)
    return ds.prefetch(AUTOTUNE), names

# ░░ 7 · Inference ░░
probs_list, y_true, class_names, branches = [], None, None, []
for m in TOP_MODELS:
    print(f"\n==> Inferring with member: {m}")
    prep,size = REG[m]
    print(f"    using batch size: {batch_for(size)}")
    ds, class_names = build_ds(DATA_ROOT, size, prep, batch_for(size))
    if y_true is None:
        y_true = np.concatenate([y for _,y in ds])
    path = RESULTS_ROOT/m/"ckpts"/"best.keras"
    if not path.exists():
        path = RESULTS_ROOT/m/f"fossil_{m}.keras"
    model = tf.keras.models.load_model(path, compile=False)
    probs_list.append(model.predict(ds, verbose=1).astype("float32"))
    branches.append((model,size,prep))
    gc.collect()

probs = np.stack(probs_list, -1)            # (N,C,M)
W = class_w(TOP_MODELS, class_names, mix=0.35)
ens = (probs * W[np.newaxis,:,:]).sum(-1)
ens /= ens.sum(1, keepdims=True)

# ░░ 8 · Report ░░
yt, yp = y_true.argmax(1), ens.argmax(1)
print("\n--- Class-Aware Weighted Ensemble Report (In-Memory) ---")
print(classification_report(yt, yp, target_names=class_names, digits=4))

report_str = classification_report(yt, yp, target_names=class_names, digits=4)

out_path = RESULTS_ROOT / "classification_report_ensemble_top-2.txt"
out_path.parent.mkdir(parents=True, exist_ok=True)  # just in case

with open(out_path, "w", encoding="utf-8") as f:
    f.write("--- Class-Aware Weighted Ensemble Report ---\n")
    f.write(report_str + "\n")

cm = confusion_matrix(yt, yp)
plt.figure(figsize=(14,12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
plt.title("Confusion Matrix – Top-2 Ensemble")
plt.tight_layout(); plt.show()

# ░░ 9 · Deployable model ░░
@register_keras_serializable(package="Fossil")
class ClassAwareEnsemble(tf.keras.Model):
    def __init__(self,names,branches,sizes,W,**kw):
        super().__init__(**kw)
        self.br   = [b for b,_,_ in branches]
        self.sz   = sizes
        self.prep = [tf.keras.layers.Lambda(p) for _,_,p in branches]
        self.W    = tf.Variable(W, trainable=False, dtype="float32")

    def call(self,x,training=False):
        x = tf.map_fn(crop_to_bbox, x)
        x = tf.image.grayscale_to_rgb(tf.image.rgb_to_grayscale(tf.cast(x,self.compute_dtype)))
        outs = [m(prep(tf.image.resize(x, sz)), training=False)
                for m, sz, prep in zip(self.br, self.sz, self.prep)]
        y = tf.stack(outs, -1) * tf.cast(self.W, outs[0].dtype)[tf.newaxis,:,:]
        y = tf.reduce_sum(y, -1)
        return y / tf.reduce_sum(y, -1, keepdims=True)

sizes = [s for _,s,_ in branches]
ensemble = ClassAwareEnsemble(TOP_MODELS, branches, sizes, tf.constant(W))
_ = ensemble(tf.zeros([1,*CANONICAL_SIZE,3]))
ensemble.save(DEPLOY_FILE)
print("\n✅ Saved top-2 ensemble to", DEPLOY_FILE)
```
